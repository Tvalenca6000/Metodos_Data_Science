```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]

titulo = "Gráfico X e Y"
eixox = "Eixo X"
eixoy = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

# Gráfico X e Y
plt.plot(x1,y1)
plt.show()
```

```python
plt.title("Gráfico de Barras")

# Gráfico de Barras
plt.bar(x1,y1)
plt.show()
```

```python
plt.title("Gráfico de Barras 2")

plt.bar(x1,y1, label = "Grupo 1")
plt.bar(x2,y2, label = "Grupo 2")
plt.legend()
plt.show()
```

```python
plt.title("Scatterplot: gráfico")

plt.scatter(x1, y1, label="Meus Pontos", color="m",marker="h", s=75)
# Plotando com os pontos ligados
plt.plot(x1, y1, linestyle="--")

plt.legend()
plt.show()
# Para salvar a imagem:
# plt.savefig("figura1.png", dpi=72)
```

```python
# Crescimento da População Brasileira 1980 - 2016
# DataSus
dados = open("populacao_brasileira.csv").readlines()

anos = []
populacao = []

for i in range(len(dados)):
  if i != 0:
    linha = dados[i].split(";")
    anos.append(int(linha[0]))
    populacao.append(int(linha[1]))

plt.plot(anos, populacao)
plt.title("Crescimento da População Brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População em Milhões")

plt.show()
```

```python
plt.bar(anos, populacao, color="#e4e4e4")
plt.plot(anos, populacao, color="k", linestyle="--")
```

```python
# Boxplot - Diagrama de Caixa
import random

vetor = []

for i in range(100):
  numero_aleatorio = random.randint(0,10000)
  vetor.append(numero_aleatorio)

plt.boxplot(vetor)
plt.title("Boxplot")
plt.show()
```
